```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
# Let's scrape a simple webpage
URL = "https://priyanshu.com.np/genai/"

# Use the requests library to fetch the webpage content.
response = requests.get(URL)
if response.status_code == 200:
    print("Successfully fetched webpage!")
else:
    print("Failed to fetch the webpage. Status code:", response.status_code)
```

```python
# Create a BeautifulSoup object and extract specific content from the HTML page.
soup = bs(response.content, 'html.parser')

# Pretty print the HTML content
print(soup.body.prettify()) # soup.body.prettify() also works
```

```python
# Locates the first <h2> tag in the document
first_header = soup.find(["h1","h2"])
print(first_header)

# Locates all the <h2> tags in the document
headers = soup.find_all(["h1","h2"])
print(headers)
```

```python
# Pass in attributes to find/find_all function
about = soup.find_all("ul")
about
```
